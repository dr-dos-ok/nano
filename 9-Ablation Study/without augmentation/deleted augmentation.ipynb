```python
# Separate by label (assuming label is in the first column)
label_0 = df[df[0] == 0]
label_1 = df[df[0] == 1]

# Keep only the first 250 rows of label 0
label_0_reduced = label_0.head(250)

# Concatenate with all label 1 rows
final_df = pd.concat([label_0_reduced, label_1], ignore_index=True)

# Save to new CSV (no header, no index)
final_df.to_csv('dataset-segmented-trimmed.csv', header=False, index=False)
```
